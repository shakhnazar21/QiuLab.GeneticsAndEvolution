```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("codonaa.csv")
```

```python
def rnaToProtein(rnaLength, rf, ):
  rna = np.random.choice(["A","C","U","G"],size = rnaLength)
  rna="".join(rna)
  protein = ""
  for j in range(0,rnaLength - 2,3):
    if (rf + j + 3) <= rnaLength:
      codon = rna[rf+j:rf+j+3]
      for i in range(df.shape[0]):
        if df.iloc[i,0] == codon:
          a = df.iloc[i,1]
          protein = "".join([protein, a])
  return(protein)
```
